```python
df = df.drop(columns=['F22','F23','F24','F25','F26','F27','F28','F29','F41'])
```

```python
X = df.drop("Class", axis=1); y = df["Class"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_val, model.predict(X_val))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_val, model.predict(X_val)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)
```

```python
accuracy_score(y_val, rf.predict(X_val))
```

```python
print(classification_report(y_val, rf.predict(X_val)))
```

```python
pd.Series(rf.feature_importances_, index=X_train.columns).sort_values(ascending=False).head(15)
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(n_estimators=500, learning_rate=0.05, max_depth=6, random_state=42, eval_metric='logloss')
xgb.fit(X_train, y_train)
```

```python
accuracy_score(y_val, xgb.predict(X_val))
```
